```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.font_manager as fm
font_name = fm.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
plt.rc('font', family=font_name)
```

```python
df=pd.read_csv('ulsanp.csv')
df
```

```python
df = df.filter(['승인일자','거래금액'])
df
```

```python
df=df.groupby('승인일자').sum()
df
```

```python
bydate=df['거래금액'].sort_index()
bydate
```

```python
df['거래금액'].plot(kind="bar", grid=True, figsize=(20,5))
plt.ylabel('거래금액')
plt.title('일자별 거래금액')
```

```python
fig,ax0 = plt.subplots(figsize=(15,5))
ax1 = ax0.twinx()

ax0.bar( bydate.index, bydate.values, label='일자별 거래금액',color='r')
ax0.set_title('일자별 거래금액')
ax0.set_xlabel('승인일자')
ax0.set_ylabel("거래금액", color = 'red')
ax0.legend(loc='upper left')
ax1.plot( bydate.index, bydate.values.cumsum(), label='누적 일자별 거래금액',color='g')
ax1.set_ylabel("누적 거래금액",color='green')
ax1.legend(loc='upper right')
plt.show()
```
