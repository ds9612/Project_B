```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.font_manager as fm
font_name = fm.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
plt.rc('font', family=font_name)
```

```python
df=pd.read_csv('ulsanp.csv')
df
```

```python
df = df.filter(['승인일자','거래금액'])
df
```

```python
df=df.groupby('승인일자').sum()
df
```

```python
df=df.groupby('승인일자').cumsum()
df
```

```python
df['거래금액'].plot(.cumsum(), grid=True, figsize=(20,5))
plt.ylabel('거래금액')
plt.title('일자별 거래금액')
```
