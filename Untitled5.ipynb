```python
df=pd.read_csv('ulsanp.csv')
df
```

```python
df = df.drop('가맹점ID', axis = 1)
df = df.drop('가맹점명', axis = 1)
df = df.drop('가맹점주(가입자)', axis = 1)
df = df.drop('고객명', axis = 1)
df = df.drop('고객 생년월일', axis = 1)
df = df.drop('고객 성별', axis = 1)
df = df.drop('승인번호', axis = 1)
df = df.drop('승인시간', axis = 1)
df = df.drop('카테고리', axis = 1)
df
```

```python
df.dtypes
```

```python
df.groupby(['승인일자'])['거래금액'].sum()
```

```python
df['승인일자'] = pd.to_datetime(df['승인일자'], format='%Y%m%d', errors='raise')
```

```python
df.dtypes
```

```python
df['date'] = df['승인일자'].dt.date
df['month'] = df['승인일자'].dt.year.astype(str) +"-"+ df['승인일자'].dt.month.astype(str)
df['weekdays'] = df['승인일자'].dt.dayofweek

df[['date', 'month', 'weekdays']]
```

```python
weekday_map= {0:'월', 1:'화',  2:'수', 3:'목', 4:'금', 5:'토', 6:'일'}
df['weekday'] = df['weekdays'].apply(lambda x : weekday_map[x])

df[['month', 'weekdays', 'weekday','승인일자', '거래금액']]
```

```python
plt.figure(figsize=(20,5))
plt.title('요일별 거래금액')
sns.countplot(data=df.sort_values(by='weekdays'), x="weekday")
```

```python
plt.figure(figsize=(20,5))
plt.title('요일별 거래금액')
sns.countplot(data=df.sort_values(by="date",ascending=True), x="month")
```
